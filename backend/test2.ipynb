```python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline, BitsAndBytesConfig
```

```python
model_id = "microsoft/Phi-3-mini-4k-instruct"

# 1. This configuration is the "Secret Sauce" for your 4GB GPU
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_use_double_quant=True,
    bnb_4bit_compute_dtype=torch.float16
)
```

```python
tokenizer = AutoTokenizer.from_pretrained(model_id)
```

```python
# 2. Load the model with the quantization config

model = AutoModelForCausalLM.from_pretrained(
    model_id,
    device_map="auto",
    trust_remote_code=True,
    quantization_config=bnb_config, # This compresses the model to ~2.2GB
    attn_implementation="eager"
)

# 3. Use the pipeline (Same as Colab)
pipe = pipeline("text-generation", model=model, tokenizer=tokenizer)

messages = [{"role": "user", "content": "Explain AI to me."}]

# Note: use_cache=False is a workaround for the error you saw earlier
output = pipe(messages, max_new_tokens=200, return_full_text=False, use_cache=False)
print(output[0]['generated_text'])
```

```python
from llama_cpp import Llama

# 1. Initialize the model
# n_gpu_layers=-1 means "put as many layers as possible on the GPU"
llm = Llama(
    model_path="./gemma-2b-it.Q4_K_M.gguf",
    n_gpu_layers=-1, 
    n_ctx=2048,      # Context window
    verbose=False    # Reduces console clutter
)

# 2. Create the prompt using Gemma's specific format
prompt = "<start_of_turn>user\nExplain AI in simple words.<end_of_turn>\n<start_of_turn>model\n"

# 3. Generate response
output = llm(
    prompt,
    max_tokens=200,
    stop=["<end_of_turn>"],
    echo=False
)

# 4. Print result
print(output["choices"][0]["text"])
```
